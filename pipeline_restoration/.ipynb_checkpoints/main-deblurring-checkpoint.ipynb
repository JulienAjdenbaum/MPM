# 2) Estimate noise parameters and prepare rescaling matrix

**Pour Jaj :** dans cette partie on estime le niveau de bruit. On fait l'hypothse d'un bruit de Poisson qu'on approxime par un bruit Gaussien dit "heteroscedastique" (i.e. de variance qui varie avec l'intensit du pixel). Si tu veux plus d'info tu peux regarder l'article qu'on est en train d'crire avec Claire (je te l'avais envoy ?).
En gros 
$$ y = Hx + N(0, a Hx + b)$$
avec $a$ et $b$ deux constantes  determiner. 

Deux possibilits pour toi :
1) Estimer les paramtres $a$ et $b$ en executant le code ci-dessous

2) Fixer $a = 0$ et $b = \sigma^2$

Tant que tu travailles sur des donnes simules j'opterais pour 2).

```python
def estimate_params(y, params=None, plot_regression=False):
    if params is not None:
        n3_min, n3_max, z_init, z_inter = params
        im = y[:, :, n3_min : n3_max]
    else :
        im = y

    im_conv = blur(im, im.shape)

    xq, list_S, d, r = LM_quantifier(im_conv[:, :, :], 8)
    
    list_var = []
    list_I = []
    for S in list_S:
        n = np.sum(S)
        I = np.sum(S*im)/n
        list_I.append(I)
        list_var.append(1/(n-1) * np.sum((S*(im - I))**2))

    a, b = scipy.stats.linregress(list_I[:], list_var[:])[:2]
    
    if plot_regression == True:
        plt.figure()
        plt.plot(list_I[:-1], list_var[:-1], 'ro', label='Donnes')
        plt.plot(list_I[:-1], a * np.array(list_I[:-1]) + b, 'b-', label="Droite de rgression")
        plt.xlabel(r'$\hat{z}$', fontsize=20)
        plt.ylabel(r'$\sigma^2$', fontsize=20)
        plt.legend()
        plt.show()
    
    return a, b

a, b  = estimate_params(y, plot_regression=True)
```

```python
print("Estimated parameters:")
print("a = ", a)
print("b = ", b)
```

# 3) Restoration

## Load PSF kernel

```python
h = np.real(sio.loadmat('Data/muscle_gdf8/profondeur/kernel_psf.mat')['h']) # size 21 x 21 x 21

# define LinearOperator x --> h * x
H = generate_blur(h, (n1, n2, n3))
```

## Set parameters of the restoration algorithm

```python
x_0 = np.copy(y)           # initialization 
delta = 0.1                # parameter in the TV regularization 
eta = 1.0                  # parameter on the constraint. Should always be close to 1 !
x_min = 0                  # minimum pixel value in the restored image
x_max = 1                  # maximum pixel value in the restored image
res = (0.049,0.049,0.1)    # resolution
max_iter = 1000            # maximum number of iteration of the algorithm
tolerance = 2*1e-04*N    # for stopping criterion
noise_params = (a, b)
```

## Set scaling matrix D

```python
y_conv =  blur(y, y.shape, KERNEL_SIZE=11, sigma=5)
D_scale = np.ones(N) / np.sqrt(abs(a * y_conv.reshape(N) + b))
```

## Run algorithm 

**Pour Jaj**:
L'algo est constitu de deux algos embriqus l'un dans l'autre. Quand on execute l'algo, il y a les itrations affiches. 

"iter" (en minuscule) --> algo intrieur

"ITER" (en majuscule) --> algo exterieur

C'est pas une ide incroyable je sais, il faudra que je change cela un jour. Mon critre d'arret porte sur le truc qui s'appelle "criterion" dans l'affichage. Ici il correspond  la diffrence entre deux itres de l'algo exterieur. Mais entre nous ce n'est pas idal. 

Il y a beaucoup de quantits affiches : 
- Psi : fonction de rgularisation
- R1 : fonction de pnalisation pour la contrainte 1 (qui est que les pixels doivent tre entre 0 et 1)
- R2 : fonction de pnalisation pour la contrainte 2 (qui porte sur le terme d'attache aux donnes, et qui est controle par le paramtre eta)


```python
x_opt =  P_MMS(np.copy(y), H, delta, eta, x_0, np.copy(y), x_min, x_max, res, noise_params, D_scale, max_iter, tolerance)
```
